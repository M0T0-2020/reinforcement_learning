```python
class Environment:
    def get_state(self, env, _obs):
        env_low = env.observation_space.low 
        env_high = env.observation_space.high 
        # normalize
        state = (_obs - env_low)/(env_high - env_low)
        return state.tolist()

    def get_reward(self, env, _obs, reward):
        reward = float(_obs[0])
        #reward = float(reward)
        #env_low = env.observation_space.low 
        #env_high = env.observation_space.high 
        # normalize
        #state = (_obs - env_low)/(env_high - env_low)
        return reward
    def get_action(self,env, action):
        action = int(action)

        return action

    def get_data(self, env, obs, next_obs, reward, action):
        state = self.get_state(env, obs)
        next_state = self.get_state(env, next_obs)
        low = env.observation_space.low[0]
        high = env.observation_space.high[0]
        reward = float((next_obs[0]-low)/(high-low))
        reward = 2*reward-1
        
        if next_obs[0]>=env.observation_space.high[0]-0.2:
            reward*=10
        #reward = self.get_reward(env, reward)
        action = self.get_action(env, action)

        return state, action, next_state, reward
```

```python
#make env
env = gym.make('MountainCar-v0')
```

```python
action_space = env.action_space.n
output_size = 1
initial_stddev=0.3
param_noise = AdaptiveParamNoiseSpec(initial_stddev=initial_stddev, desired_action_stddev=0.2, adaptation_coefficient=1.01)
agent = Agent(action_space=action_space, output_size=1,  
              critic_lr=0.001, actor_lr=0.002, environment=Environment(),
              tau=0.001, param_noise=param_noise)
```

```python
# 楽観的初期化
#loss_list = agent.init_param_optim(100, env)
#plt.plot(loss_list)
```

```python
def end_episode(episode, total_reward, rewards):
    rewards.append(total_reward)
    # doneがTrueになったら１エピソード終了
    s = 50
    if episode%s == 0:
        if episode==0:
            mean_total_reward = total_reward
        else:
            m = episode//s
            mean_total_reward = np.mean(rewards[(m-1)*s:m*s])
        print(f'episode: {episode}, total_reward　mean: {mean_total_reward}')
    return rewards
```

```python
actions = []
rewards = []
critic_loss_list = []
actor_loss_list = []
stddev_list = []

# 10000エピソードで学習する
for episode in range(300):
    total_reward = 0
    observation = env.reset()
    for _ in range(2000):
        env.render()
        action =  agent.get_action(env, observation)
        actions.append(action)
        stddev_list.append(agent.param_noise.current_stddev)
        next_observation, reward, done, _ = env.step(action)
        total_reward += reward
        critic_loss, actor_loss = agent.update_agent(env, observation, next_observation, reward, action)
        if not(critic_loss==-1 and actor_loss==-1):
            critic_loss_list.append(critic_loss)
            actor_loss_list.append(actor_loss)
            
        observation = next_observation
        if observation[0]>=env.observation_space.high[0]-0.001:
            rewards = end_episode(episode, total_reward, rewards)
            break
    rewards = end_episode(episode, total_reward, rewards)
```

```python
critic_loss_df = pd.DataFrame(critic_loss_list, columns=['loss'])
actor_loss_df = pd.DataFrame(actor_loss_list, columns=['loss'])
reward_df = pd.DataFrame(rewards, columns=['reward'])
action_df = pd.DataFrame(actions, columns=['action'])

critic_loss_df['loss_r30'] = critic_loss_df['loss'].rolling(30,min_periods=1).mean()
actor_loss_df['loss_r30'] = actor_loss_df['loss'].rolling(30,min_periods=1).mean()
reward_df['reward_r30'] = reward_df['reward'].rolling(30,min_periods=1).mean()
action_df['action'] = action_df['action'].map({0:"push left", 1:"no push", 2:"push right"})
```
